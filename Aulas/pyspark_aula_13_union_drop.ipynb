```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = (
    SparkSession.builder.appName("pyspark_aula_12")
    .config("spark.sql.repl.eagerEval.enabled", True)
    .getOrCreate()
)
```

```python
df = spark.read.parquet("./bases/LIVROS.parquet").limit(5)
```

## Duplicando o dataFrame df

```python
dup = df.union(df)
dup
```

```python
df1 = df.select('cnpj_editora', 'data_lancamento', 'ean', 'isbn10', 'numero_paginas', 'preco','id')
df1
```

```python
df.union(df1)
```

```python
df.unionByName(df1)
```

```python
df2 = df1.withColumn('Quantidade', F.lit(1))
df2
```

```python
df2.unionByName(df1, allowMissingColumns=True)
```
