```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = (
    SparkSession.builder.appName("pyspark_aula_12")
    .config("spark.sql.repl.eagerEval.enabled", True)
    .getOrCreate()
)
```

```python
df = spark.read.parquet("./bases/LOGINS.parquet")
```

```python
df.groupBy('estado').agg(F.count('*').alias('contagem'))
```

```python
livros = spark.read.parquet("./livros.parquet").alias('livros')

compras = spark.read.parquet("./compras.parquet").alias('compras')
```

```python
df2 = compras.join(livros, compras.cd_livro == livros.id, 'inner')
```

```python
df2.select('livros.*')
```
